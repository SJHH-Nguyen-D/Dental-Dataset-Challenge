## Data Preprocessing

### Notes

to impute for these missing values we should have a strategy for each type:

* most_frequent for integer/frequency based features and categorical features

* mean for float data type features

```python
for i in cols_with_missing.columns:
    print(i, "", cols_with_missing[i].dtype)
```

```python
from sklearn.impute import SimpleImputer

imp = SimpleImputer(missing_values=np.nan, strategy="most_frequent")

# univariate imputing for freq/int columns
freq_cols = ["num_claims", 
             "total_num_serv_lines", 
             "num_claims_dep", 
             "total_num_serv_lines_dep",
             "ftr_employee_student_presence_flag",
             "ftr_employee_spouse_dependent_flag",
             "ftr_employee_children_dependent_flag",
             "ftr_employee_handicap_presence_flag",
             "ftr_employee_male_presence_flag",
             "ftr_employee_female_presence_flag",
             "MARKET_SEGMENT_ADJ",
             "cde_marital_status_eeeb",
             "ftr_employee_retire_flag"]

df[freq_cols] = pd.DataFrame(imp.fit_transform(cols_with_missing[freq_cols]), 
                             columns=freq_cols)
```

```python
# univariate imputing for float columns
mean_imp = SimpleImputer(missing_values=np.nan, strategy="mean")

missing_float_cols = [i for i in set(cols_with_missing.columns).difference(set(freq_cols))]

df[missing_float_cols] = pd.DataFrame(mean_imp.fit_transform(cols_with_missing[missing_float_cols]), 
             columns=missing_float_cols)
```

```python
# proof that we've imputed for all missing values

df.isnull().sum().sort_values(ascending=False)
```
